```python
import re
import random
from tqdm.notebook import tqdm
from collections import Counter
from datasets import load_dataset, Dataset
```

```python
# use HuggingFace datasets library to load an Arabic text data
# For more information, visit this link:
# https://huggingface.co/datasets/arabic_billion_words
data_dict = {}
data_dict['SaudiYoum_text'] = load_dataset('arabic_billion_words', 'SaudiYoum', split="train")
data_dict['Ryiadh_text'] = load_dataset('arabic_billion_words', 'Ryiadh', split="train")
```

```python
# This function simply remove any none Arabic characters from the the given text
def remove_non_arabic(text):
    return ' '.join(re.sub(u"[^\u0621-\u063A\u0641-\u064A ]", " ", text,  flags=re.UNICODE).split())
```

```python
# Loop over all datasets that you loaded and
# for each dataset, loop over each line to clean it.
# Also, count the frequency of each word in the corpus
# to use it later to select the top N words

counter = Counter()
for data_name, data in data_dict.items():
    with open(f'Clean_{data_name}.txt', 'w') as f1:
      for text in tqdm(data['text']):
          text = remove_non_arabic(text)
          if text != '':
              counter.update(text.split())
              _=f1.write(f'{text.strip()}\n')

with open('vocab_list.txt', 'w') as f1:
  for k, v in tqdm(counter.items()):
    _=f1.write(f'{k}\t{v}\n')
```
